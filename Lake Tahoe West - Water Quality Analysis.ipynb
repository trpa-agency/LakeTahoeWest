## Calculate Hectares, Acres, Total Sediment Yield, Total Phosphorus Yield, and Total Fine Sediment Yield.

### For Current Condition

```python
import arcpy, os
from arcpy import env
from arcpy.sa import *

# set environment vairables
env.overwriteOutput=True
env.extent = "MAXOF"
env.workspace ="C:\\GIS\\PROJECT\\LakeTahoeWest\\Data\\LANDIS\\WaterQualityAnalysis\\WaterQualityAnalysis_Harvest.gdb"

# function to check if field exists already
def FieldExist(featureclass, fieldname):  
    fieldList = arcpy.ListFields(featureclass, fieldname)  
    fieldCount = len(fieldList)  
    if (fieldCount == 1):  
        return True  
    else:  
        return False

# fields to check for 
field1 = "TotalYield_Sediment"
field2 = "TotalYield_Phosphorus"
field3 = "Area_Hectares"
field4 = "Area_Acres"
# generate list of feature classes
fcs = arcpy.ListFeatureClasses("lt_*", "ALL")

for fc in fcs:
    # get feature class name
    desc = arcpy.Describe(fc)
    name = desc.baseName
    #add field for total yield
    if (FieldExist(fc, field1) == False):  
        arcpy.AddField_management(fc, "TotalYield_Sediment", "DOUBLE", "", "", "", "TotalYield_Sediment", "NULLABLE", "REQUIRED")
    else:
        pass
    if (FieldExist(fc, field2) == False):  
        arcpy.AddField_management(fc, "TotalYield_Phosphorus", "DOUBLE", "", "", "", "TotalYield_Phosphorus", "NULLABLE", "REQUIRED")
    else:
        pass
    if (FieldExist(fc, field3) == False):  
        arcpy.AddField_management(fc, "Area_Hectares", "DOUBLE", "", "", "", "Area_Hectares", "NULLABLE", "REQUIRED")
    else:
        pass
    if (FieldExist(fc, field4) == False):  
        arcpy.AddField_management(fc, "Area_Acres", "DOUBLE", "", "", "", "Area_Acres", "NULLABLE", "REQUIRED")
    else:
        pass
    # calculate hectares
    arcpy.CalculateField_management(fc, "Area_Hectares", "!shape.area@HECTARES!", "PYTHON3", "")
    print ("Calculated Hectares of: " + desc.baseName)
    # calculate acres
    arcpy.CalculateField_management(fc, "Area_Acres", "!shape.area@ACRES!", "PYTHON3", "")
    print ("Calculated Acres of: " + desc.baseName)
    # calculate yields
    with arcpy.da.UpdateCursor(fc, [field1, field2, field3, field4, 'SedYld_kgh', 'TP_kg_ha_']) as cur:
        for row in cur:
            row[0] = row[2] * row[4]
            row[1] = row[2] * row[5]
            cur.updateRow(row)
        print("Calculated total sediment and phosphorus yield for: " + name)
```
